```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a signal that is a sum of two sinusoids with frequencies 100 Hz and 150 Hz
t = np.linspace(0, 1, 1000, endpoint=False)
x = np.sin(2*np.pi*100*t) + np.sin(2*np.pi*150*t)

# Compute the magnitude spectrum using FFT with a rectangular window
X = np.abs(np.fft.fft(x))

# Plot the magnitude spectrum
freqs = np.fft.fftfreq(len(x), 1/1000)
plt.plot(freqs[:len(freqs)//2], X[:len(freqs)//2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()

# Identify the frequencies with the highest magnitudes
idx_max = np.argsort(X[:len(freqs)//2])[-2:]
freq_max = freqs[idx_max]
print('Frequencies of highest peaks:', freq_max, 'Hz')

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a signal that is a sum of two sinusoids with frequencies 100 Hz and 150 Hz
t = np.linspace(0, 1, 1000, endpoint=False)
x = np.sin(2*np.pi*100*t) + np.sin(2*np.pi*150*t)

# Apply a Kaiser-Bessel window with shape parameter = 10 and side lobe attenuation = -30 dB
window = np.kaiser(len(x), 10)
x_win = x * window

# Compute the magnitude spectrum using FFT
X = np.abs(np.fft.fft(x_win))

# Plot the magnitude spectrum
freqs = np.fft.fftfreq(len(x), 1/1000)
plt.plot(freqs[:len(freqs)//2], X[:len(freqs)//2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.show()

# Identify the frequencies with the highest magnitudes
idx_max = np.argsort(X[:len(freqs)//2])[-2:]
freq_max = freqs[idx_max]
print('Frequencies of highest peaks:', freq_max, 'Hz')
```
